Plot Line plot with formatting to display two studentâ€™s data with last 4 semester SPI. also use legend in plot , different line styles and marker characters.

```python
import matplotlib.pyplot as plt
```

```python
sem = [1, 2, 3, 4]

stud1 = [7, 8, 8.4, 5]
stud2 = [8.4, 8.37, 9, 7]
```

```python
plt.plot(sem, stud1, label="rahul", linestyle="--")
plt.plot(sem, stud2, label="krunal", linestyle="-.")

plt.xlabel("sem")
plt.ylabel("SPI")

plt.legend(loc=2, facecolor="black", labelcolor="white")
plt.show()
```
