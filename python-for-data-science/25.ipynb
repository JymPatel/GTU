Write a python program for plotting of geographical data using basemap.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Create a Basemap instance
m = Basemap(
    projection='merc',
    llcrnrlat=-80, urcrnrlat=80,
    llcrnrlon=-180, urcrnrlon=180,
    resolution='c'
)

# Draw coastlines, countries, and states
m.drawcoastlines()
m.drawcountries()
m.drawstates()

# Draw parallels and meridians
m.drawparallels(np.arange(-90, 90, 30), labels=[1,0,0,0])
m.drawmeridians(np.arange(-180, 180, 60), labels=[0,0,0,1])

# Plot some data points
lats = [37.77, 34.05, 40.71]
lons = [-122.42, -118.24, -74.01]
x, y = m(lons, lats)
m.scatter(x, y, marker='o', color='red', label='Cities', zorder=5)

# Set the plot title
plt.title("Geographical Data Plot with Basemap")

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
