Write a python program to do line plotting, scatter plotting, Bar chart, box plot ,
histogram and pie chart.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5,6]
y = [10, 20, 12, 23, 34, 53]
plt.plot(x, y, color="red")
plt.title("Line Plot", fontsize=15)
plt.show()
```

```python
plt.scatter(x, y)
plt.title("Scatter Plot")
```

```python
plt.bar(x, y, width=0.3)
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x,labels = ["1st"],patch_artist = True)
plt.show()
```

```python
x = np.random.randint(20,180,(50))
plt.hist(x, color="green", edgecolor="red")
plt.title("Histogram Plot")
plt.xlabel("Number")
plt.ylabel("Repetition")
plt.show()
```

```python
import random

x = [random.randint(0, 100) for i in range(6)]
y = ["Pizza", "Burger", "Sushi", "Pav Bhaji", "Salad", "Lasangna"]

plt.pie(x, labels=y, colors=["r", "b", "g", "y", "m", "c"])
plt.title("Most loved food!")
plt.show()
```
